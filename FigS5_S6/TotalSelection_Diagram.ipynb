```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
M=1
K=500
alpha=5
xs=[100,500,2000]
ys=np.arange(0,510,10)
plt.figure(figsize=(2,2))
for x in xs:
    phi1=alpha*ys/(K+x+alpha*ys)
    plt.plot(ys,phi1,linewidth=5)
plt.tick_params(direction='in')
plt.xlabel('plasmid size',fontsize=16)
plt.ylabel('phenotypic selection',fontsize=16)
plt.ylim(0,1)
plt.savefig('TotalSelection_Diagram_1.pdf')
plt.show()
```

```python
M=1
K=500
alpha=5
xs=[100,500,2000]
ys=np.arange(0,510,10)
plt.figure(figsize=(2,2))
for x in xs:
    phi2=x/(x+alpha*ys)
    plt.plot(ys,phi2,linewidth=5)
plt.tick_params(direction='in')
plt.xlabel('plasmid size',fontsize=16)
plt.ylabel('phenotypic selection',fontsize=16)
plt.ylim(0,1)
plt.savefig('TotalSelection_Diagram_2.pdf')
plt.show()
```

```python
plt.figure(figsize=(2.5,2))
x=2000
ys=np.arange(0,1000,100)
alphas=np.arange(0.5,5.5,0.5)
mm=np.zeros((len(ys),len(alphas)))
for i in range(len(ys)):
    for j in range(len(alphas)):
        mm[i,j]=alphas[j]*ys[i]/(K+x+alphas[j]*ys[i])
ax=sns.heatmap(data=mm,cmap='OrRd',linewidths=0.5,linecolor='gray')
ax.invert_yaxis()
plt.tick_params(direction='in',left=False,bottom=False)
plt.xlabel(r'$\alpha$',fontsize=16)
plt.ylabel('plasmid size',fontsize=16)
plt.savefig('TotalSelection_Diagram_3.pdf')
plt.show()
```

```python
M=1
K=500
alpha=10
xs=[100,500,1000]
ys=np.arange(0,510,10)
plt.figure(figsize=(2,2))
for x in xs:
    phi1=alpha*ys/(K+x+alpha*ys)
    phi2=x/(x+alpha*ys)
    plt.plot(ys,phi1+phi2,linewidth=5)
plt.tick_params(direction='in')
plt.xlabel('plasmid size',fontsize=16)
plt.ylabel('total selection',fontsize=16)
plt.ylim(0.5,1)
plt.savefig('TotalSelection_Diagram_5.pdf')
plt.show()
```

```python
M=1
K=1000
alphas=[5,10,20]
xs=np.arange(100,10100,100)
plt.figure(figsize=(2,2))
for alpha in alphas:
    ym=np.sqrt(K*xs+xs**2)/alpha
    plt.plot(xs,ym,linewidth=5)
plt.tick_params(direction='in')
plt.xlabel('chromosome size',fontsize=16)
plt.ylabel('optimal plasmid size',fontsize=16)
plt.xscale('log')
plt.yscale('log')
plt.savefig('TotalSelection_Diagram_7.pdf')
plt.show()
```
